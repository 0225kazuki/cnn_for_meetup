```python
import sqlite3
import glob
import matplotlib.pyplot as plt
from PIL import Image
% matplotlib inline
```

CREATE TABLE train_image_urls(
  "ImageID" TEXT,
  "Subset" TEXT,
  "OriginalURL" TEXT,
  "OriginalLandingURL" TEXT,
  "License" TEXT,
  "AuthorProfileURL" TEXT,
  "Author" TEXT,
  "Title" TEXT,
  "OriginalSize" TEXT,
  "OriginalMD5" TEXT,
  "Thumbnail300KURL" TEXT
);
CREATE TABLE train_image_labels(
  "ImageID" TEXT,
  "Source" TEXT,
  "LabelName" TEXT,
  "Confidence" TEXT
);
CREATE TABLE class(
  "LabelName" TEXT,
  "description" TEXT
);

```python
conn = sqlite3.connect("google_open_images.db")
cur = conn.cursor()
cnt = 0 


for name in glob.glob("google_open_images/images/*"):
    
    fig = plt.figure()
    org_img = Image.open(name)
    plt.imshow(org_img)
    plt.show()
    
    image_id = name.split("/")[-1].split(".")[0]
    print(image_id)
    cur.execute("""select LabelName from train_image_labels where ImageID = '{}'""".format(image_id))
    for id_ in cur.fetchall():
        id_ = id_[0]
        cur.execute("""select description from class where LabelName = '{}'""".format(id_))
        print(cur.fetchall()[0],end="")
    print()
    if cnt > 10 :
        break
    else:
        cnt+=1

conn.close()
```
