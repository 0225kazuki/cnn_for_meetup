```python
conn = sqlite3.connect("google_open_images.db")
cur = conn.cursor()
cur.execute("""select ImageID,Thumbnail300KURL from train_image_urls limit 100""")


with open("over1000labelname.txt",'r') as f:
    label_name = f.readline().strip()
    while label_name:
        print(label_name)
        cur.execute("""select train_image_urls.ImageID,train_image_urls.Thumbnail300KURL from train_image_labels inner join train_image_urls on train_image_labels.ImageID = train_image_urls.ImageID where train_image_labels.LabelName = '{}' """.format(label_name))
        with open("over1000urls.txt",'a') as g:

            cnt = 0
            for image_id, url in cur.fetchall():
                if cnt > 999:
                    break
                if url:
                    g.write(image_id+','+url+'\n')
                    cnt += 1
                else:
                    pass
        label_name =  f.readline().strip()

conn.close()
```

```python
# 全画像取得 最初の100件

import urllib.request

conn = sqlite3.connect("google_open_images.db")
cur = conn.cursor()
cur.execute("""select ImageID, from train_image_urls limit 100""")

data = cur.fetchall()

conn.close()
# 画像ファイルのURLを開く
# （urlに画像ファイルのURLを指定）
for ID, url in data:
    print("url",url)
    try:
        request = urllib.request.urlopen(url)
    except:
        print("error\t",ID,"\t",url)
        continue
    
    # ファイルをバイナリモードで開き、URLの内容を書き込み
    # （file_nameに保存時のファイル名を指定）
    f = open('google_open_images/images/{}.jpg'.format(ID), "wb")
    f.write(request.read())

    # ファイルを閉じる
    f.close()
```

```python
# 全画像取得 次の100件から5万件とってくる

import urllib.request
from tqdm import tqdm_notebook as tqdm

conn = sqlite3.connect("google_open_images.db")
cur = conn.cursor()

for cnt in range(1,500):
    print("fetch ",100*cnt,"-",100*(cnt+1),' data')
    
    cur.execute("""select ImageID,Thumbnail300KURL from train_image_urls limit {},100""".format(100*cnt))

    data = cur.fetchall()

    # 画像ファイルのURLを開く
    for ID, url in tqdm(data):
        try:
            request = urllib.request.urlopen(url)
        except:
            print("error\t",ID,"\t",url)
            continue

        # ファイルをバイナリモードで開き、URLの内容を書き込み
        f = open('google_open_images/test_images/{}.jpg'.format(ID), "wb")
        f.write(request.read())

        # ファイルを閉じる
        f.close()
        
conn.close()
```
